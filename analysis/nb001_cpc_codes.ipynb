```python
import pandas as pd
```

```python
cpc_codes = pd.read_csv(Config.input_dir + "/cpc-codes/titles.csv")

subclass_df = cpc_codes[(~cpc_codes["class"].isnull()) & (~cpc_codes["subclass"].isnull()) & (cpc_codes["group"].isnull())].reset_index(drop=True)
subclass_df["code"] = subclass_df["code"].apply(lambda x: x[:3])

codes = []
titles = []
for key, group in subclass_df.groupby("code"):
    title = " ".join(group["title"].to_list())
    codes.append(key)
    titles.append(title)

subclass = pd.DataFrame({"code": codes, "subclass_title": titles})

group_df = cpc_codes[(~cpc_codes["class"].isnull()) & (~cpc_codes["subclass"].isnull()) & (~cpc_codes["group"].isnull())].reset_index(drop=True)
group_df["code"] = group_df["code"].apply(lambda x: x[:3])

codes = []
titles = []
for key, group in group_df.groupby("code"):
    title = " ".join(group["title"].to_list())
    codes.append(key)
    titles.append(title)

group = pd.DataFrame({"code": codes, "group_title": titles})

cpc_codes_df = subclass.merge(group, on=["code"], how="left")
```
